# Visualizing Earnings Based On College Majors

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their Github repo.

Using visualizations, we can start to explore questions from the dataset like:
1. Do students in more popular majors make more money?
2. How many majors are predominantly male? Predominantly female?
3. Which category of majors have the most students?

# Method One: A grouped bar plot to compare the number of men with the number of women in each category of majors.

```python

Major_category_unique = recent_grads["Major_category"].unique().tolist()
#Major_category_unique = set(recent_grads["Major_category"])

def Major_category_gender(gender):
    counts_gender = []
    for i in Major_category_unique:
        count = 0  
        for j in range(recent_grads.shape[0]):
            if recent_grads.iloc[j]["Major_category"] == i:
                count += recent_grads.iloc[j][gender]
        counts_gender.append(count)
    return counts_gender
          
Major_category_Men = Major_category_gender("Men")
Major_category_Female = Major_category_gender("Women")


# Major_category_unique = pd.Series(data = Major_category_unique)
```

```python
d = {'Major_category': Major_category_unique, 'Men': Major_category_Men, 'Women': Major_category_Female}
df = pd.DataFrame(data=d)
```

```python
#recent_grads.plot.bar(x ="Major_category", y = ["Men","Women"],figsize=(20,10))
```

```python
df.plot.bar(x ="Major_category", y = ["Men","Women"],figsize=(20,10))
```

# Method One: A grouped bar plot to compare the number of men with the number of women in each category of majors. (Using groupby)

```python
df_category_sum = recent_grads.groupby(['Major_category'],as_index=False).sum()
```

```python
df_category_sum
```

```python
df_category_sum.plot.bar(x ="Major_category", y = ["Men","Women"],figsize=(20,10))
```
